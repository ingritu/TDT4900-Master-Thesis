### Imports

```python
import numpy as np
from pathlib import Path
import pandas as pd
from sklearn.utils import shuffle
```

```python
ROOT_PATH = Path().absolute().parent
ROOT_PATH
```

```python
para_path = ROOT_PATH.joinpath('data', 'external', 'p5.csv')
```

```python
coco_path = ROOT_PATH.joinpath('data', 'processed', 'karpathy_split', 'c5_train_clean.csv')
```

```python
SEED = 12345
np.random.seed(SEED)
```

```python
para_df = pd.read_csv(para_path)
para_df
```

```python
coco_df = pd.read_csv(coco_path)
coco_df
```

```python
SAMPLE_SIZE=20
```

```python
shuff = shuffle(para_df)
shuff = shuff.reset_index(drop=True)
p_sample = shuff.loc[:SAMPLE_SIZE-1, ['caption_id', 'caption']]
p_sample
```

```python
## find the orignial captions in coco_df
sample_d = {}
for i in range(len(p_sample)):
    caption_id = p_sample.loc[i, 'caption_id']
    caption = p_sample.loc[i, 'caption']
    sample_d[caption_id] = caption
```

```python
for i in range(len(coco_df)):
    image_name = coco_df.loc[i, 'image_name']
    cap_id = coco_df.loc[i, 'caption_id']
    cap = coco_df.loc[i, 'caption']
    # locate the paraphrase
    candidate = cap_id+'#p'
    if candidate in sample_d:
        print('OG:\t' + str(cap) + '\tPARA:\t' + str(sample_d[candidate]))
    
```
